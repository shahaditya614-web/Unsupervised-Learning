```python
# ===============================================
# Unsupervised Learning
# Unlabeled Data
# K-Means
# Customer Segmentation
# Clustering
# ===============================================

# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Step 2: Create Unlabeled Customer Data
customer_data = {
    "Annual_Income": [20,22,24,26,28,35,38,41,44,47,
                      55,58,61,64,67,75,78,81,84,87],
    "Spending_Score": [25,30,35,40,45,60,65,70,75,80,
                       20,25,30,35,40,85,88,90,92,95]
}

df = pd.DataFrame(customer_data)

print("========== Customer Data ==========\n")
print(df)

# Step 3: Customer Data Graph (Before Clustering)

plt.figure(figsize=(8,6))

plt.scatter(df["Annual_Income"],
            df["Spending_Score"],
            color="blue",
            s=120)

plt.title("Customer Data (Unlabeled Data)")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score")
plt.grid(True)

plt.show()

# Step 4: Select Features
X = df[["Annual_Income","Spending_Score"]]

# Step 5: Apply K-Means Clustering
kmeans = KMeans(n_clusters=4, random_state=10)

df["Cluster"] = kmeans.fit_predict(X)

# Step 6: Customer Segmentation Result

print("\n========== Customer Segmentation Result ==========\n")

print(df)

# Step 7: Cluster Graph

plt.figure(figsize=(8,6))

plt.scatter(df["Annual_Income"],
            df["Spending_Score"],
            c=df["Cluster"],
            cmap="rainbow",
            s=120)

plt.scatter(kmeans.cluster_centers_[:,0],
            kmeans.cluster_centers_[:,1],
            color="black",
            marker="X",
            s=250,
            label="Centroids")

plt.title("Customer Segmentation using K-Means")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score")
plt.legend()
plt.grid(True)

plt.show()

# Step 8: Cluster Centers

print("\n========== Cluster Centers ==========\n")

print(kmeans.cluster_centers_)

# Step 9: Number of Customers in Each Cluster

print("\n========== Customers in Each Cluster ==========\n")

print(df["Cluster"].value_counts().sort_index())
```
